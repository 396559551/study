```python
#     linewidth,color,marker,
#     markersize,markeredgecolor,
#     markerfactcolor,label,alpha)
# x：指定折线图的x轴数据；
# y：指定折线图的y轴数据；
# linestyle：指定折线的类型，可以是实线、虚线、点虚线、点点线等，默认文实线；
# linewidth：指定折线的宽度
# marker：可以为折线图添加点，该参数是设置点的形状；

# markeredgecolor：设置点的边框色；
# markerfactcolor：设置点的填充色；
# label：为折线图添加标签，类似于图例的作用；
# 导入模块
import pandas as pd
import matplotlib.pyplot as plt

# 设置绘图风格
plt.style.use('ggplot')
# 设置中文编码和负号的正常显示
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
#读出数据
article_reading=pd.read_csv('./exercise_data/wechart.csv')
article_reading.date = pd.to_datetime(article_reading.date)
print(article_reading.date)
#取出8月份至9月28号的数据
sub_data=article_reading.loc[article_reading.date>='2017-08-1',:]
#设置图框大小
fig=plt.figure(figsize=(10,6))
#绘图
plt.plot(sub_data.date,#x轴数据
         sub_data.article_reading_cnts,#y轴数据
         linestyle='-',#绘线风格
         lw=2,
         color='steelblue',
         marker='o',#点的形状

         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown'# 点的填充色
         )
#设置标签和标题
plt.xlabel('时间')
plt.ylabel('阅读人数')
plt.title('阅读情况')
## 为了避免x轴日期刻度标签的重叠，设置x轴刻度自动展现，并且45度倾斜
fig.autofmt_xdate(rotation = 45)
plt.show()
```

```python
#只需要在代码中写入两次plot函数即可，
import matplotlib as mpl
#设置图框大小
plt.figure(figsize=(10,6))
#人数趋势
plt.plot(sub_data.date,
         sub_data.article_reading_cnts,
         linestyle='-',
         marker='o',
         lw=2,
         color='steelblue',
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown'# 点的填充色
         )
#人次趋势
plt.plot(sub_data.date,
         sub_data.article_reading_times,
         linestyle='-',
         marker='o',
         lw=2,
         color='red',
         markeredgecolor='blue', # 点的边框色
         markerfacecolor='g'# 点的填充色
         )
#添加标签和标题
plt.xlabel('时间')
plt.ylabel('阅读人数')
plt.title('人数和人次趋势')


# 为了避免x轴日期刻度标签的重叠，设置x轴刻度自动展现，并且45度倾斜
fig.autofmt_xdate(rotation = 45)
#显示图例
plt.legend()
plt.show()
```
