```python
#中国的四个直辖市分别为北京市、上海市、天津市和重庆市，其2017年上半年的GDP分别为12406.8亿、
#13908.57亿、9386.87亿、9143.64亿。对于这样一组数据，我们该如何使用条形图来展示各自的GDP水平呢？
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import  numpy as np
#中文乱码
plt.rcParams['font.sans-serif'] =['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#中国的四个直辖市分别为北京市、上海市、天津市和重庆市，
# 其2017年上半年的GDP分别为12406.8亿、
#13908.57亿、9386.87亿、9143.64亿。对于这样一组数据，我们该如何使用条形图来展示各自的GDP水平呢？

GDP=[12406.8,13908.57,9386.87,9143.64]
plt.bar(range(4),GDP,alpha=0.8,align = 'center',color='steelblue')
#添加y标签
plt.ylabel('GDP')
#y轴的范围
plt.ylim([500,15000])
#x轴的标签
plt.xticks(range(4),['北京市','上海市','天津市','重庆市'])
#设置标题
plt.title('四大城市比拼')
#为每个条形图添加数值标签
for x,y in enumerate(GDP):
    plt.text(x,y+100,'%s'%round(y,1),ha='center')
plt.show()
```

```python
#很多人在买一本书的时候，都比较喜欢货比三家，
#python数据分析实战》在亚马逊、当当网、中国图书网、京东和天猫的
# 最低价格分别为39.5、39.9、45.4、38.9、33.34。
# 针对这个数据，我们也可以通过条形图来完成，
# 这里使用水平条形图来显示：
price=[39.5,39.9,45.4,38.9,33.34]
plt.barh(range(5),price,align='center',alpha=0.8,color='red')
#设置x的标签
plt.xlabel='价格'
#设置x的范围
plt.xlim([20,50])
#设置y的刻度标签
plt.yticks(range(5),['亚马逊','当当网','中国图书网','京东','天猫'])
#设置标题
plt.title('价格大作战')
#为每个条形图添加数值标签
for x,y in enumerate(price):
    plt.text(y+0.1,x,'%s' %y,va='center')
plt.show()
```

```python
# 胡润财富榜：亿万资产超高净值家庭数
# 利用水平交错条形图
# 对比2016年和2017年亿万资产超高净值家庭数（top5），
# 其数据如下
#       北京      上海      香港      深圳      广州
# 2016 15600    12700     11300       4270     3620
# 2017 17400    14800     12000       5200     4020

y2016=[15600,12700,11300,4270,3620]
y2017=[17400,14800,12000,5200,4020]
city=['北京','上海','香港','深圳','广州']
bar_width=0.35

plt.bar(np.arange(5), y2016, label = '2016', color = 'steelblue', alpha = 0.8, width = bar_width)
plt.bar(np.arange(5)+bar_width, y2017, label = '2017', color = 'indianred', alpha = 0.8, width = bar_width)

#设置y刻度
plt.ylim([2000,20000])
#设置y标签
plt.ylabel('亿万资产超高净值家庭数')
#设置x标签
#plt.xlabel('Top5城市')
#设置x刻度标签
plt.xticks(np.arange(5),city)
#标题
plt.title('亿万财富家庭数Top5城市分布')
## 显示图例
plt.legend()
#为每个条形图添加数值标签
for a2016,b2016 in enumerate(y2016):
    plt.text(a2016,b2016+100,'%s' %y2016)
for a2017,b2017 in enumerate(y2017):
    plt.text(a2017+bar_width,b2017+100,'%s' %y2017,ha='center')

plt.show()
```
