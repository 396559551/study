```python
# x：指定要绘制直方图的数据；
# bins：指定直方图条形的个数；
# range：指定直方图数据的上下界，默认包含绘图数据的最大值和最小值；
# normed：是否将直方图的频数转换成频率；
# weights：该参数可为每一个数据点设置权重；
# cumulative：是否需要计算累计频数或频率；
# bottom：可以为直方图的每个条形添加基准线，默认为0；
# histtype：指定直方图的类型，默认为bar，除此还有’barstacked’, ‘step’, ‘stepfilled’；
# align：设置条形边界值的对其方式，默认为mid，除此还有’left’和’right’；
# orientation：设置直方图的摆放方向，默认为垂直方向；
# rwidth：设置直方图条形宽度的百分比；
# log：是否需要对绘图数据进行log变换；
# color：设置直方图的填充色；
# label：设置直方图的标签，可通过legend展示其图例；
# stacked：当有多个数据时，是否需要将直方图呈堆叠摆放，默认水平摆放；
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import norm
# 中文和负号的正常显示
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#整体乘客的年龄直方图
titanic=pd.read_csv('./titanic/train.csv')
#检查有无年龄的缺失
titanic.Age.isnull()
#删除缺失的
titanic.dropna(subset=['Age'],inplace=True)
```

```python
plt.style.use('ggplot')
#绘图
plt.hist(titanic['Age'],#绘图数据
         bins=20,#指定条数
         color='steelblue',#指定填充色
         edgecolor='k',#直方图边界色
         )
plt.title('直方图')
#显示图例
plt.legend('直方图')
# 去除图形顶部边界和右边界的刻度
plt.tick_params(top='off', right='off')
#画出
plt.show()
```

```python
# 绘图：乘客年龄的累计频率直方图
plt.hist(titanic.Age,
         bins=np.arange(titanic.Age.min(),titanic.Age.max(),5),
         density=True,#是否以密度的形式显示
         color = 'steelblue', # 指定填充色
         edgecolor = 'k',
         )
#设置坐标轴标签和标题
plt.title('频率直方图')
plt.xlabel('年龄')
plt.ylabel('累计频率')
# 显示图例
plt.legend(loc = 'best')
plt.show()
```

```python
plt.hist(titanic.Age,
         bins=np.arange(titanic.Age.min(),titanic.Age.max(),5),
         density=True,
         color='steelblue',
         edgecolor='k')
#设置坐标标签和标题
plt.title('正态分布图')
plt.xlabel('年龄')
plt.ylabel('频率')
plt.legend()
#生成正态曲线数据
x1=np.linspace(titanic.Age.min(),titanic.Age.max(),1000)
normal = norm.pdf(x1, titanic.Age.mean(), titanic.Age.std())
#画出正态分布曲线
line1=plt.plot(x1,normal)
plt.legend([line1],'正态分布',loc='best')
plt.show()
```

```python
#我们可以通过两个hist将不同性别的直方图绘制到一张图内
# 提取不同性别的年龄数据
age_female = titanic.Age[titanic.Sex == 'female']
age_male = titanic.Age[titanic.Sex == 'male']

# 设置直方图的组距
bins = np.arange(titanic.Age.min(), titanic.Age.max(), 2)
# 男性乘客年龄直方图
plt.hist(age_male, bins = bins, label = '男性', color = 'steelblue', alpha = 0.7)
# 女性乘客年龄直方图
plt.hist(age_female, bins = bins, label = '女性', alpha = 0.6)

# 设置坐标轴标签和标题
plt.title('乘客年龄直方图')
plt.xlabel('年龄')
plt.ylabel('人数')

# 去除图形顶部边界和右边界的刻度
plt.tick_params(top='off', right='off')

# 显示图例
plt.legend()
# 显示图形
plt.show()
```
